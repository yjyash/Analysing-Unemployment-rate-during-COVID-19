```python
from dash import Dash,dcc,html,Input,Output,dash_table
import plotly.express as px
import pandas as pd
```

```python
df=pd.read_csv('Ue.csv')
#df=px.data.gapminder()
```

```python

app=Dash(__name__)
```

```python
 #create drop down buttons
app.layout=html.Div([
    dash_table.DataTable(data=df.to_dict('records'),page_size=10),
    html.H4('Unemployment during Covid-19'),
    html.P('Select Graph'),
    dcc.Dropdown(
        id='selection',
        options=['UNemployment rate ',"Estimated Unemployment Rate(%) Vs Estimated Employed" ,' Estimated Labour Participation Rate (%)','  Estimated Labour Participation Rate (%) vs  Region.1','labour participant rate with date with region wise','easty','percentage change in unemployment in each state after lockdown' ],
        value='UNemployment rate ',
    ),

dcc.Loading(dcc.Graph(id='graph'),type='cube')
])

@app.callback(
    Output("graph","figure"),
    Input('selection','value'),
)

def dispaly_animated_graph(selection):
    animations= {
        "UNemployment rate ":px.scatter(
            df,x='Region',
         y=' Estimated Employed',
        #     # #animation_frame='year',
        #     # animation_group='country',
        #     # size='pop',
        color=' Estimated Unemployment Rate (%)',
        #     # hover_name='country',
        ),
        "Estimated Unemployment Rate(%) Vs Estimated Employed":px.scatter(
            df,x= ' Estimated Unemployment Rate (%)',
            y=' Estimated Employed',
            color="Region.1",
        #     animation_group='country',
        ),
        " Estimated Labour Participation Rate (%)":px.scatter(
            df,x= ' Estimated Labour Participation Rate (%)',
            y=' Estimated Employed',
            # color="Region",
        #     animation_group='country',
        ),
        "  Estimated Labour Participation Rate (%) vs  Region.1":px.histogram(
            df,x= 'Region.1' ,
            y=' Estimated Labour Participation Rate (%)',
            color="Region",
            # animation_group='Frequency',
        ),
        " Estimated Labour Participation Rate (%) with estimated employed":px.scatter(
            df,x= ' Estimated Labour Participation Rate (%)',
            y=' Estimated Employed',
            color="Region",
        #     animation_group='country',
        ),
       "labour participant rate with date with region wise":px.scatter_matrix(
           df,template='plotly',
    dimensions=[' Estimated Labour Participation Rate (%)',
                ' Date'],
    color='Region'),
    "easty":px.box(
        df,x='Region',
        y=' Estimated Unemployment Rate (%)',
        color='Region.1',
        title='Unemployment rate',template='plotly'
# fig.update_layout(xaxis={'categoryorder':'total descending'}
 ),
 "percentage change in unemployment in each state after lockdown":px.bar(data,
     x='Region',
     y=' Estimated Employed',
     color=' Estimated Unemployment Rate (%)',
            title='percentage change in Unemployment in each state after lockdown',template='ggplot2'
)
#  "impact of lockdown(map):"px.scatter_geo(df,'longitude', 'latitude', color="Region",
#                      hover_name="Region", size=" Estimated Unemployment Rate (%)",
#                      animation_frame=" Date",scope='asia',template='plotly',title='Impack of lockdown on employement across regions')

# fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 2000

# fig.update_geos(lataxis_range=[5,35], lonaxis_range=[65, 100],oceancolor="#6dd5ed",
#     showocean=True)

        
         
    }
    return animations[selection]
```
